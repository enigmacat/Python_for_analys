## Задание 2

На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

> **kwargs** - это способ передачи произвольного количества аргументов в функцию в виде словаря. В функции kwargs представляется как словарь, в котором ключами являются названия аргументов, а значениями - сами аргументы.  
> Таким образом, для решения данной задачи, необходимо:  
> - Создать функцию, которая будет принимать произвольное количество аргументов в виде kwargs  
> - Просуммировать количество каждого фрукта и вернуть общее количество фруктов на складе.

````python
# использование цикла for
def count_fruits_1(**kwargs):
    total_count = 0
    for fruit, count in kwargs.items():
        total_count += count
    return total_count
````

````python
print('Общее количество фруктов на складе =', count_fruits_1(
    груши=167, яблоки=234, бананы=345, мандарины=457, апельсины=897, абрикосы=123
    ))
````

````python
# использование функции reduce из модуля functools, которая принимает 
# функцию и последовательность и возвращает единственное значение.
from functools import reduce

# используем функцию reduce, которая принимает анонимную функцию lambda, 
# которая складывает все значения словаря kwargs и возвращает общее количество фруктов на складе.
def count_fruits_2(**kwargs):
    return reduce(lambda x, y: x + y, kwargs.values())
````

````python
print('Общее количество фруктов на складе =', count_fruits_2(
    груши=167, яблоки=234, бананы=345, мандарины=457, апельсины=897, абрикосы=123
))
````

````python
# используем comprehensions для решения данной задачи.
# используем генератор списков, который возвращает список значений словаря kwargs, а затем суммируем все значения.
def count_fruits_3(**kwargs):
    return sum([count for count in kwargs.values()])
````

````python
print('Общее количество фруктов на складе =', count_fruits_3(
    груши=167, яблоки=234, бананы=345, мандарины=457, апельсины=897, абрикосы=123
))
````

## Задание 3

Дан список с затратами на рекламу.  
Но в данных есть ошибки, некоторые затраты имеют отрицательную величину.  
Удалите такие значения из списка и посчитайте суммарные затраты  
[100, 125, -90, 345, 655, -1, 0, 200]  
Используйте list comprehensions

````python
# используем list comprehensions. Создаём новый список, исключив из него отрицательные значения. 
# Вычисляем сумму всех элементов полученного списка.
expenses = [100, 125, -90, 345, 655, -1, 0, 200]
positive_expenses = [x for x in expenses if x >= 0]
total_expenses = sum(positive_expenses)

print('Список без отрицательных сумм: ', positive_expenses)
print('Суммарные затраты =', total_expenses)
````

### 4.1 Найдите, какая выручка у компании в ноябре
Используйте list comprehensions

````python
# создать список, содержащий только те элементы из списка с суммами покупок, 
# которые соответствуют ноябрю. Это можно сделать с помощью генератора списков и 
# метода split для разбиения даты на составляющие:
november_sales = [sale for date, sale in zip(
    date, purchases_by_dates) if date.split('-')[1] == '11']

# мы проходимся по списку дат и сумм покупок параллельно, используя функцию zip. 
# Для каждой пары элементов мы проверяем, соответствует ли месяц даты ноябрю, и если да, 
# то добавляем соответствующую сумму покупки в новый список november_sales.

november_revenue = sum(november_sales)

print('Выручка компании в ноябре =', november_revenue)
````

### 4.2 Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

````python
def transform_months(d: list, r: list) -> dict:
    result = {}
    m_name = ['январь', 'февраль', 'март', 'апрель', 'май', 'июнь',
              'июль', 'август', 'сентябрь', 'октябрь', 'ноябрь', 'декабрь']
    m_txt = ['-01-', '-02-', '-03-', '-04-', '-05-', '-06-',
             '-07-', '-08-', '-09-', '-10-', '-11-', '-12-']
    for m in range(12):
        r_s = sum(purchases_by_dates[i]
                  for i in range(len(r)) if m_txt[m] in d[i])
        if r_s != 0:
            result.update({m_name[m]: r_s})
    return (result)


print('Выручка по месяцам:', transform_months(date, purchases_by_dates))
````

````python
# использование модуля pandas, который предоставляет более мощные средства для работы с данными.
import pandas as pd

# Эта функция создает объект DataFrame из двух списков и преобразует столбец date в объект datetime. 
# Затем добавляется столбец month, содержащий месяц и год в формате мм.гггг. 
# Затем функция использует метод groupby для группировки данных по месяцам и суммирует выручку по месяцам 
# с помощью метода sum. Наконец, результат преобразуется в словарь с помощью метода to_dict.

def revenue_by_month(dates: list[str], revenue: list[int]) -> dict[str, int]:
    df = pd.DataFrame({'date': dates, 'revenue': revenue})
    df['date'] = pd.to_datetime(df['date'])
    df['month'] = df['date'].dt.strftime('%B %Y')
    return df.groupby('month')['revenue'].sum().to_dict()

print('Выручка по месяцам:', revenue_by_month(date, purchases_by_dates))
````
